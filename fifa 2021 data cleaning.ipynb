### Column: [Name]
### For the Name field we have two columns: Name and LongName
- The Name column contains the abbreviated first name followed by the last name
- The LongName column contains the full name but contains special characters which we have to remove
- We see that the column playerUrl contains the complete name of the player without any special characters so we can use this column to extract the full name of players and store in a new column 'Full Name'

```python
def extract_player_name(playerurl):
    name = playerurl.split('/')[5].replace('-', ' ')
    return name.title()
    
# Apply the function to the 'playerUrl' column and create new column for Full Name
fifa.insert(data.columns.get_loc('Name') + 1, 'Full Name', data['playerUrl'].apply(extract_player_name))
```

### Column: [Contract]
### This shows the player's contract details

```python
# Determine the data type for the column
fifa['Contract'].dtype
```

#### Creating additional columns 'Contract Start', 'Contract End' and 'Contract Length(years)' from the "Contract" column
- The Contract columns have values of 3 types which are date ranges like '2003 ~ 2023' , 'Jun 30, 2023 On Loan' and 'Free' so we can 
- Utilize these to determine the 'Contract Start', 'Contract End' and 'Contract Length(years)' for each player

```python
# define function extract_contract_info to extract values from the Contract column
def extract_contract_info(contract):
    # If the value is Free or On Loan set the start and end date to nan and contract length to 0
	if contract == 'Free' or 'On Loan' in contract:
		start_date = pd.NA
		end_date = pd.NA
		contract_length = 0
    # If the start and end date is present then extract the values
	else:
		start_date,end_date = contract.split(' ~ ')
		start_year = int(start_date[:4])
		end_year = int(end_date[:4])
		contract_length = end_year - start_year
	return start_date, end_date, contract_length

# Apply the function to Contract Column & create new columns
new_cols = ['Contract Start', 'Contract End' , 'Contract Length(years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
	fifa.insert(loc = fifa.columns.get_loc('Contract')+1+i, column = new_cols[i],value=new_data[i])
```

#### Verify the newly created columns 'Contract Start', 'Contract End' and 'Contract Length(years)'

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Length(years)']].sample(10)
```

#### Create new column [Contract Status]
- The new column 'Contract Status' will help us determine the current status of contract for any player

```python
def categorize_contract_status(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'

# Add the column for Contract Status
fifa.insert(fifa.columns.get_loc('Contract Length(years)')+1, 'Contract Status', fifa['Contract'].apply(categorize_contract_status))
```

#### Verify the newly created column [Contract Status]

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Length(years)', 'Contract Status']].sample(5)
```

### Column: [Height]
### This shows the player's height

```python
# Determine the data type for the column
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

#### Convert the heights into cm and also convert the data type to integer

```python
def convert_height(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches*2.54)

# Apply the height function to Height column 
fifa['Height'] = fifa['Height'].apply(convert_height)
fifa['Height'].unique()
```

#### Rename the column [Height] to [Height(cm)] since all the values are now in cm

```python
fifa = fifa.rename(columns={'Height' : 'Height(cm)'})
fifa.sample(5)
```

### Column: [Weight]
### This shows the player's weight

```python
# Determine the data type for the column
fifa['Weight'].dtype
```

```python
fifa['Weight'].unique()
```

#### Convert the weight into km and also convert the data type to integer

```python
def convert_weight(weight):
    # Check if weight is already an integer
    if isinstance(weight, int):
        return weight
    else:
        if "kg" in weight:
            return int(weight.strip("kg"))
        else:
            weight_lbs = int(weight.strip("lbs"))
            return round(weight_lbs/2.20)

# Apply the weight function to Weight column 
fifa['Weight'] = fifa['Weight'].apply(convert_weight)
fifa['Weight'].unique()
```

#### Rename the Weight column to Weight(kg) since all the values are now in kg

```python
fifa = fifa.rename(columns={'Weight' : 'Weight(kg)'})
fifa.sample(5)
```

### Column: [Value], [Wage], and [Release Clause]
- The values in all the three columns are in euros(€) which needs to be formatted
- Additionally the values contains 'K' and 'M' denoting thousand and million which needs to be updated
- The column name needs to be formatted for denoting values in currency

```python
# Check the data type
fifa[['Value', 'Wage', 'Release Clause']].dtypes
```

```python
# Print the data before making changes
fifa[['Value', 'Wage', 'Release Clause']].sample(5)
```

```python
def clean_value_wage_release(value):
    # Remove the euro symbol
    value = str(value).strip("€")
    if 'K' in value:
        return float(value.replace('K', '')) * 1e3
    elif 'M' in value:
        return float(value.replace('M', '')) * 1e6
    else:
        return float(value)

# Apply the function to the columns
fifa['Value'] = fifa['Value'].apply(clean_value_wage_release)
fifa['Wage'] = fifa['Wage'].apply(clean_value_wage_release)
fifa['Release Clause'] = fifa['Release Clause'].apply(clean_value_wage_release)
```

```python
fifa = fifa.rename(columns={'Value' : 'Value(€)', 'Wage' : 'Wage(€)', 'Release Clause' : 'Release Clause(€)'})
fifa[['Value(€)', 'Wage(€)', 'Release Clause(€)']].sample(5)
```

### Column: [Weak Foot(W/F)], [SM], and [IR]
#### This shows the data of [Weak Foot(W/F)], [SM], and [IR] for each player on a scale of 1 to 5
- The ratings contains a special character(★) which needs to be removed
- Finally the data type is changed to integer

```python
fifa[['W/F', 'SM', 'IR']].dtypes
```

```python
# Print the data before making changes
fifa[['W/F', 'SM', 'IR']].sample(5)
```

```python
# Clean the columns and convert to integer
fifa[['W/F', 'SM', 'IR']] = fifa[['W/F', 'SM', 'IR']].apply(lambda x: x.str.replace('★', '').str.strip().astype(int))
fifa[['W/F', 'SM', 'IR']].sample(5)
```

### Column: [Hits]
### This shows the Hits for each player which denotes how many times the player has been searched for

```python
fifa['Hits'].dtype
```

```python
fifa['Hits'].sample(10)
```

#### Convert the Hits column into integer type clean the data

```python
def player_hits(hits):
    hits = str(hits).strip()
    if "K" in hits:
        return int((float(hits.strip("K"))*1000))
    else:
        return int(hits)

# Apply the weight function to Weight column 
fifa['Hits'] = fifa['Hits'].fillna(0)
fifa['Hits'] = fifa['Hits'].apply(player_hits)
fifa['Hits'].sample(10)
```
